```python
from openai import OpenAI
from tqdm import tqdm
import random
from sklearn.metrics import f1_score
import re
import os

client = OpenAI(
    api_key=os.getenv("MY_API_KEY"),
)

def chat_gpt(prompt, model):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content": """
                   You are helpfull assistant that based on context forms Knowledge graph. By representing as node (A, B, C) some facts ideas other entities and by edges interconnections between them.
                    Structure it this way:
                    "
                    Nodes:
                    A - > entity
                    B - > entity
                    C - > entity

                    Edges
                    A - B (desription of this edge)
                    "
                   
                   Form responses in one line
                   Example:
                   Nodes: A - > Sarah (Competitive swimmer training for national championships), B - > Training Regimen (Daily swimming, resistance training, diet), C - > Daily Swimming (5,000 meters), D - > Resistance Training (Three times a week), E - > Diet (Rich in protein and complex carbohydrates), F - > National Championships (Event for competitive swimmers). Edges: A - B (Sarah follows a strict training regimen), B - C (Training regimen includes daily swimming), B - D (Training regimen includes resistance training), B - E (Training regimen includes a carefully planned diet), C - F (Daily swimming helps prepare Sarah for national championships), D - F (Resistance training supports performance in national championships), E - F (Diet provides necessary nutrients for competition preparation)."""}, {"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content.strip()
```

```python
import pandas as pd

model = 'gpt-4o-mini'

input_csv = 'natural_prompts.csv'
prompts_df = pd.read_csv(input_csv, sep=";")
results = []
```

```python
prompts_df
```

```python
output_csv = 'result_natural_graphs.csv'
if RERUN_RESULTS:
    for index, row in tqdm(prompts_df.iterrows(), total=prompts_df.shape[0], desc="Processing rows"):
        prompt = row['prompt']
        prompt[:prompt.find("Guard")]
        try:
            result = chat_gpt(prompt, model).lower()
            second_part = prompt[prompt.find("Guard"):]

            result = 'Context: ' + result + '. '  + second_part
            results.append({
                'based_on': prompt,
                'prompt': result + second_part
            })
        
        except Exception as e:
            print(f"An error occurred with model {model}: {e}")

    results_df = pd.DataFrame(results)
    results_df.to_csv(output_csv, index=False)
    print(f"Responses have been saved to {output_csv}")
else:
    results_df = pd.read_csv(output_csv)
```
